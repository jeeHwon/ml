```python
import pandas as pd

filename = '/content/drive/My Drive/Colab Notebooks/EverybodyDL/data/data-01-test-score.csv'

xy = pd.read_csv(filename, header=None)
xy.head()
```

```python
x_data = xy.iloc[:, 0:-1]
y_data = xy.iloc[:, -1]
```

```python
print(x_data, "\nx_data shape:", x_data.shape)
print(y_data, "\ny_data shape:", y_data.shape)
```

```python
import tensorflow as tf

tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3, activation='linear'))

tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(learning_rate=1e-5))

history = tf.model.fit(x_data, y_data, epochs=2000)
```

```python
print("Your score will be ", tf.model.predict([[100, 70, 101]]))
print("Other scores will be ", tf.model.predict([[60, 70, 110], [90, 100, 80]]))
```
